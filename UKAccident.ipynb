# UK Accident History

## Data Import

### Package Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import CSV files

```python
accdnt_2005 = pd.read_csv('accidents_2005_to_2007.csv')
accdnt_2009 = pd.read_csv('accidents_2009_to_2011.csv')
accdnt_2012 = pd.read_csv('accidents_2012_to_2014.csv')
accdntTraffic = pd.read_csv('ukTrafficAADF.csv')
```

#### Columns specification

```python
print(accdnt_2005.columns)
print(accdnt_2009.columns)
print(accdnt_2012.columns)
print(accdntTraffic.columns)
```

#### Merge year dataset to one

```python
accdnt_All = pd.concat([accdnt_2005,accdnt_2009,accdnt_2012], ignore_index=True)
#accdnt_All.head()
print(accdnt_All.shape)
print(accdnt_2005.shape)
print(accdnt_2009.shape)
print(accdnt_2012.shape)
```

### Data Exploration

#### Urban vs Rural

```python
UrbanRural = accdnt_All['Urban_or_Rural_Area'].value_counts()
UrbanRural.plot.pie(figsize=(10,10))
plt.show()
```

```python
sns.barplot(x=UrbanRural.index, y=UrbanRural.values)
plt.xlabel('Rural or Urban')
plt.ylabel('Accident count')
plt.show()
```
