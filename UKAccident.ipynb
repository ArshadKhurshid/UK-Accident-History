#### Creating features and Predictive value

```python
accdnt_All.drop('Junction_Detail', axis=1)
accdnt_All.fillna('Unknown')
```

```python
PolArr_raw = accdnt_All['Did_Police_Officer_Attend_Scene_of_Accident']
Feature_raw = accdnt_All.drop(['Did_Police_Officer_Attend_Scene_of_Accident','Accident_Index','Date',
                               'Time','Location_Easting_OSGR','Location_Northing_OSGR','Year',
                              '1st_Road_Number','2nd_Road_Number','Longitude','Latitude','LSOA_of_Accident_Location','Junction_Detail'], axis=1)
```

```python
Feature_raw.dtypes
```

#### Data Preprocessing

##### One HotEncoder

```python
# One hot encode using pandas.dummies
Feature_final = pd.get_dummies(Feature_raw)

#Encode Police Arrival data to numerical value
PolArr_final = pd.get_dummies(PolArr_raw, drop_first=True)

encoded = list(Feature_final.columns)

print("{} total features after one hot encoding" .format(len(encoded)))
```

#### Shuffle and Split

```python
#Import train test split
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Feature_final, 
                                                    PolArr_final, 
                                                    test_size = 0.2, 
                                                    random_state = 0)

#Show the result for split
print("Training set has {} as samples" .format(X_train.shape[0]))
print("Testing set has {} as samples" .format(X_test.shape[0]))
print("Training y set has {} as samples" .format(y_train.shape[0]))
print("Testing y set has {} as samples" .format(y_test.shape[0]))
```

#### Performance Calculation

```python
from sklearn.metrics import fbeta_score, accuracy_score, confusion_matrix
def claf_performance(test_val, pred_val, model):
    from sklearn.metrics import fbeta_score, accuracy_score, confusion_matrix
    
    tn, fp, fn, tp = confusion_matrix(test_val, pred_val).ravel()
    acc = accuracy_score(test_val, pred_val)
    fbeta = fbeta_score(test_val, pred_val, beta=0.5)
    print("Total number of {} are True Negative".format(tn))
    print("Total number of {} are False Positive".format(fp))
    print("Total number of {} are False Negative".format(fn))
    print("Total number of {} are True Positive".format(tp))
    print("{}".format(model))
    print("Accuracy Score for model: {:.3f}".format(acc))
    print("Beta Score for model for beat=0.5: {:.3f}".format(fbeta))
    return
```

### Benchmark Model

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

claf = LogisticRegression()
claf.fit(X_train, y_train)

clf_pred =claf.predict(X_test)
claf_performance(y_test, clf_pred, claf)
```

### Model Evaluation

#### Grid Search Using Kfold

```python
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import make_scorer

def GridSearch_Classifier(classifier,parameters, splits_count, X_train, y_train):
    
    scorer = make_scorer(fbeta_score, beta=0.5)
    kfold = KFold(n_splits=splits_count, shuffle=True, random_state=0)
    
    grid_obj = GridSearchCV(classifier,parameters, scoring = scorer, cv = kfold)
    
    grid_fit = grid_obj.fit(X_train, y_train)
    
    print(grid_fit.cv_results_)
    
    return grid_fit
    
```

#### Optimized Decision Tree

```python
param = {}
data_split_nbr = 10

Opt_clf = GridSearch_Classifier(claf, param, data_split_nbr, X_train, y_train )

best_clf = Opt_clf.best_estimator_
best_predictions = best_clf.predict(X_test)
claf_performance(y_test, best_predictions, best_clf)
```
